```python
# Identity Transformations (Identity)
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the image
f = cv.imread('images/emma.jpg', cv.IMREAD_GRAYSCALE)

# Create the transformation function
t = np.arange(256, dtype=np.uint8)

# Define the transformation for each range
t[0:51] = t[0:51]             
t[51:151] = np.clip(
    (155/99) * (np.arange(51, 151) - 51) + 100,
    0, 255
)
t[151:256] = t[151:256]       

# Apply the transformation to the image
g = t[f]

# Plot the original image, transformed image, and transformation function
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0].imshow(f, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original Image')
ax[1].imshow(g, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Transformed Image')
for a in ax[0:2]:
    a.axis('off')
ax[2].plot(t)
ax[2].set_ylim([0, 255])
ax[2].set_xlim([0, 255])
ax[2].set_title('Transformation Function')
plt.show()
```

## i) For gamma = 0.6

Because the issue with this image is brightness of dark parts is too low, So we need to brightens dark parts. So I used gamma =  0.6 because it is give the best one as I see.

```python
# Gamma Correction
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

gamma = 0.6
f = cv.imread('images/highlights_and_shadows.jpg')

t = np.array([(i / 255.0) ** gamma * 255 for i in np.arange(0, 256)]).astype(np.uint8)
g = cv.LUT(f, t)

fig, ax = plt.subplots(1, 2, figsize=(12, 8))
ax[0].imshow(cv.cvtColor(f, cv.COLOR_BGR2RGB))
ax[0].set_title('Original Image')
ax[1].imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))
ax[1].set_title('Gamma Correction')
for a in ax:
    a.axis('off')
plt.show()
```

# ii)

```python
# Gamma Correction with Histograms Only
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

gamma = 0.6
f = cv.imread('images/highlights_and_shadows.jpg')

# Apply gamma correction
t = np.array([(i / 255.0) ** gamma * 255 for i in np.arange(0, 256)]).astype(np.uint8)
g = cv.LUT(f, t)

# Plot the histograms of the original and corrected images
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Calculate and plot histograms for the original image
color = ('b', 'g', 'r')
for i, col in enumerate(color):
    hist_f = cv.calcHist([f], [i], None, [256], [0, 256])
    ax[0].plot(hist_f, color=col)
    ax[0].set_xlim([0, 256])
    ax[0].set_title('Histogram of Original Image')

# Calculate and plot histograms for the gamma corrected image
for i, col in enumerate(color):
    hist_g = cv.calcHist([g], [i], None, [256], [0, 256])
    ax[1].plot(hist_g, color=col)
    ax[1].set_xlim([0, 256])
    ax[1].set_title('Histogram of Gamma Corrected Image')

plt.tight_layout()
plt.show()
```

We can see thehistogram is shifted to right after gamma correction with gamma value 0.6.

This means more pixel values are being mapped to higher intensity levels, resulting in a higher concentration of pixels in the brighter range of the histogram.
