```python
# Identity Transformations (Identity)
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the image
f = cv.imread('images/emma.jpg', cv.IMREAD_GRAYSCALE)

# Create the transformation function
t = np.arange(256, dtype=np.uint8)

# Define the transformation for each range
t[0:51] = t[0:51]             
t[51:151] = np.clip(
    (155/99) * (np.arange(51, 151) - 51) + 100,
    0, 255
)
t[151:256] = t[151:256]       

# Apply the transformation to the image
g = t[f]

# Plot the original image, transformed image, and transformation function
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0].imshow(f, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original Image')
ax[1].imshow(g, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Transformed Image')
for a in ax[0:2]:
    a.axis('off')
ax[2].plot(t)
ax[2].set_ylim([0, 255])
ax[2].set_xlim([0, 255])
ax[2].set_title('Transformation Function')
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the image
f = cv.imread('images/brain_proton_density_slice.png', cv.IMREAD_GRAYSCALE)

# Create the transformation function
t = np.arange(256, dtype=np.uint8)

# Define the transformation for each range
t[0:156] = t[0:156]          
t[156:181] = 0                   
t[181:256] = t[181:256]         

# Apply the transformation to the image
g = t[f]

# Plot the original and transformed images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(f, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(g, cmap='gray')
plt.title('Transformed Image')
plt.axis('off')

plt.tight_layout()
plt.show()

# Plot the transformation function as a graph
plt.figure(figsize=(8, 4))
plt.plot(np.arange(256), t, color='blue', linewidth=2)
plt.title('Transformation Function')
plt.xlabel('Original Intensity Values')
plt.ylabel('Transformed Intensity Values')
plt.grid()
plt.xlim(0, 255)
plt.ylim(0, 255)
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.show()
```
